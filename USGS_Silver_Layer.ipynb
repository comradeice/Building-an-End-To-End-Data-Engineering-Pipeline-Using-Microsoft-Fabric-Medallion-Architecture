### Silver Medallion/ Layer Processing for Global Earthquake Events

```python
# Import the necessary libraries

from pyspark.sql.functions import col
from pyspark.sql.types import TimestampType
```

```python
# df now is a Spark DataFrame containing data from "Files/2025-03-03_earthquake_data.json"

df = spark.read.option("multiline", "true").json("Files/2025-03-03_earthquake_data.json")
```

```python
# Transform earthquake data by extracting  key attributes and renaming them for further analysis.
df = \
df.\
    select(
        'id',
        col('geometry.coordinates').getItem(0).alias('longitude'),
        col('geometry.coordinates').getItem(1).alias('latitude'),
        col('geometry.coordinates').getItem(2).alias('elevation'),
        col('properties.title').alias('title'),
        col('properties.place').alias('place_description'),
        col('properties.sig').alias('sig'),
        col('properties.mag').alias('mag'),
        col('properties.magType').alias('magType'),
        col('properties.time').alias('time'),
        col('properties.updated').alias('updated')
        )
```

```python
# Display the table
display(df)
```

```python
# For clearer datetime representation : convert 'time' and 'updated' columns from milliseconds to timestamp format.
df = df.\
    withColumn('time', col('time')/1000).\
    withColumn('updated', col('updated')/1000).\
    withColumn('time', col('time').cast(TimestampType())).\
    withColumn('updated', col('updated').cast(TimestampType()))
```

```python
# Append the data and save as the table for Silver Medallion

df.write.mode('append').saveAsTable('earthquake_events_silver')
```

```python
# Display the final transformed table
display(df)
```
